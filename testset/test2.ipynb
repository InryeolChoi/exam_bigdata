# 제2유형

```python
df = pd.read_csv("../Dataset/Parkinsons.csv")
```

```python
df.head(10)
```

```python
df = df.drop(['name'], axis=1, inplace=False)
```

```python
import sklearn.preprocessing as preprocessing
import statsmodels.api as sm

df_norm = preprocessing.minmax_scale(df)
df2 = pd.DataFrame(df_norm)
df2.columns = df.columns
```

```python
df2 = sm.add_constant(df2, has_constant='add')
df2.head(10)
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['status'], axis=1)
y = df['status'].astype('category')

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.1, random_state=42)
```

```python
model = sm.Logit(y_train, X_train)
results = model.fit(method='bfgs', maxiter=1000)
```

```python
results.summary()
```

```python
from sklearn.metrics import f1_score

def cutoff(y, threshold):
    Y = y.copy()
    Y[Y > threshold] = 1
    Y[Y <= threshold] = 0
    return (Y.astype(int))

test_y_pred_prob = results.predict(X_test)
test_y_pred = cutoff(test_y_pred_prob, 0.8)

f1_score(y_test, test_y_pred)
```

```python
test_y_pred_prob = results.predict(X_test)
test_y_pred = cutoff(test_y_pred_prob, 0.5)

f1_score(y_test, test_y_pred)
```

# 제3유형

```python
df = pd.read_csv("../Dataset/HR-Employee-Attrition.csv")
```

```python
table = pd.crosstab(df['OverTime'], df['MaritalStatus'])
print(table)
```

```python
x1 = table.loc['Yes', 'Married']
x2 = table.loc['No', 'Married']
print(int(abs(x1 - x2)))
```

```python
from scipy import stats

result = stats.chi2_contingency([table.loc['Yes'], table.loc['No']])
print(round(result.statistic, 2))
```

```python
print(round(result.pvalue, 4))
if round(result.pvalue, 4) < 0.05:
    print("H_0 기각")
else:
    print("H_0 기각 불가")
```
