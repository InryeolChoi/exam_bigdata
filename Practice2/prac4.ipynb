```python
# 2유형
bank = pd.read_csv("../Dataset/Bank_Personal_Loan_Modeling.csv")
```

```python
bank.info()
```

```python
bank.head(10)
```

```python
bank = bank.drop(labels=['ID', 'ZIP Code'], axis=1)
```

```python
bank.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

X = bank.drop(labels=['Personal Loan'], axis=1)
y = bank['Personal Loan']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state=1111, test_size=0.3)
```

```python
from sklearn.preprocessing import Normalizer
scaler = Normalizer()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

training_acc = []
test_acc = []

for i in range(1, 25):
    classifier = KNeighborsClassifier(n_neighbors = i)
    classifier.fit(X_train, y_train)

    training_acc.append(classifier.score(X_train, y_train))
    test_acc.append(classifier.score(X_test, y_test))
```

```python
test_acc
```

```python
df = pd.read_csv("../Dataset/HR-Employee-Attrition.csv")
```

```python
table = pd.crosstab(df['OverTime'], df['MaritalStatus'])
table
```

```python
from scipy.stats import chi2_contingency

chi2, p, df, expected = chi2_contingency(table)

chi2
```

```python
pval = round(p, 4)
print(pval)
if pval < 0.05:
    print("H0 기각") 
else : 
    print("H0 기각 불가")
```

```python
import pandas as pd

train = pd.read_csv("../Dataset/data_6_2_train.csv")
test = pd.read_csv("../Dataset/data_6_2_test.csv")
```

```python
train.info()
```

```python
train.head(10)
```

```python
targetVal = train.pop('구분')
```

```python
from sklearn.model_selection import train_test_split

train = pd.get_dummies(train)
test = pd.get_dummies(test)

train_X, valid_X, train_y, valid_y = train_test_split(train, targetVal, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(train_X, train_y)
```

```python
y_pred = classifier.predict(valid_X)
```

```python
from sklearn.metrics import f1_score
f1_score(valid_y, y_pred, average='macro')
```

```python
pred_test = classifier.predict(test)
result = pd.DataFrame({'pred' : pred_test})
result
```
